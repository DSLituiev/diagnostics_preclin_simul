```python
import pandas as pd
import numpy as np
from numpy import sqrt, abs

from itertools import combinations
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
step = 0.01
# p = np.arange(step, 0.5+step, step)
p = 10.0**np.arange(-3,0,)
q = p
N = np.arange(5,51)

yy = ((1-p)*(1-q))**N[:,np.newaxis]
```

### Definition
sick negative rate (_p_) -- expected proportion of negative test results in sick cohort
healthy positive rate (_q_) -- expected proportion of positive test results in the healthy cohort

### Description of experimental procedure
Test results are expected to be positive for sick cohort, except for some proportion _p_ due to both technical and biological variability. Similarly, some proportion _q_ of healthy cohort will show positive test results.

### Question
How the expected proportion of flawless features (features for which proportion of both healthy positives and sick negatives is zero) will change with increased sample size

### Model
expected proportion of sick positives (flawless): $1-p$
expected proportion of healthy negatives (flawless): $1-q$

Expected proportion of cases with two flawless disease samples: 
$$(1-p)\cdot(1-p)$$
Expected proportion of cases with $N$ flawless disease samples:
    $$(1-p)^N$$
Expected proportion of cases with $N$ flawless disease samples and $N$ flawless healthy samples:
    $$(1-p)^N \cdot (1-q)^N $$
    

```python
plt.subplots(1, figsize=(8,6))
for nn, p_ in enumerate((p)):
    plt.plot(N, yy[:,nn], label = "flip rate = 1/%u" % int(1/p_), lw=2)
# plt.ylabel("sick negative rate = healthy positive rate")
# plt.yscale("log")
# plt.xscale("log")
plt.ylabel("proportion of flawless observations")
plt.xlabel("number of individuals per group, $N$")
plt.legend(loc="lower right")
```
